```python
#### IMPORT LIBRARIES ####
import pandas as pd 
import matplotlib.pyplot as plt
import geopandas as gpd
import fiona as fn
```

```python
# load dengue dataset
data = pd.read_csv("ph_dengue.csv")
```

```python
# Only include dengue case counts for data viz
cases = data[['Month', 'Year', 'Region', 'MTD_Cases']]
cases.head()
```

```python
# load shapefile of Philippines
ph_shape = gpd.read_file("ph_shapefile/Regions.shp")
ph_shape
```

```python
# Check plot of shapefile
plt.rcParams['figure.figsize'] = [20, 10] 
ph_shape.plot()
```

```python
ph_shape.columns
```

```python
# Rename all regions for easier reference and to match with the region column in the cases dataframe
ph_shape['REGION'] = ph_shape['REGION'].apply({'Ilocos Region (Region I)': 'Region I',
 'Cagayan Valley (Region II)': 'Region II',
 'Central Luzon (Region III)': 'Region III',
 'CALABARZON (Region IV-A)': 'Region IV-A',
 'MIMAROPA (Region IV-B)': 'MIMAROPA',
 'Bicol Region (Region V)': 'Region V',
 'Western Visayas (Region VI)': 'Region VI',
 'Central Visayas (Region VII)': 'Region VII',
 'Eastern Visayas (Region VIII)': 'Region VIII',
 'Zamboanga Peninsula (Region IX)': 'Region IX',
 'Northern Mindanao (Region X)': 'Region X',
 'Davao Region (Region XI)': 'Region XI',
 'SOCCSKSARGEN (Region XII)': 'Region XII',
 'Autonomous Region of Muslim Mindanao (ARMM)': 'ARMM',
 'Cordillera Administrative Region (CAR)': 'CAR',
 'Caraga (Region XIII)': 'CARAGA',
 'Metropolitan Manila': 'NCR'}.get) 
```

```python
# Also, rename the region column
ph_shape =ph_shape.rename(columns = {'REGION': 'Region'})
ph_shape
```

```python
# Merge cases and shapefile into one dataframe for analysis
cases_map = pd.merge(cases, ph_shape, on= ['Region'])
```

```python
cases_map.head()
```

```python
## Define a function to filter dengue case counts by month and year
def filtered_counts(df, month, year):
    return df[(df['Month']==month) & (df['Year']==year)]
```

```python
## Define a function to visualize Dengue Cases 
def dengue_choropleth(filtered_dengue, select_month):
    plot_map = gpd.GeoDataFrame(filtered_dengue).plot(column='MTD_Cases', cmap='Oranges', figsize=(40,20), linewidth=0.5, k=15, edgecolor = 'black', legend=True, legend_kwds={'label': 'Dengue Case Counts'})
    plot_map.set_title('Map of Month-To-Date Dengue Cases in the Philippines',fontdict={'fontsize': 25})
    plot_map.annotate('Dengue Reported Cases from January to {} 2019'.format(select_month), xy=(0.6, .05), xycoords='figure fraction', fontsize=12, color='#555555')
    plot_map.set_axis_off()
```

```python
## Filter MTD dengue case counts for all regions reported in August 2019
filtered_dengue_case_counts = filtered_counts(cases_map, 'Aug', 2019)

## Plot choropleth map of MTD dengue case counts for August 2019 
dengue_choropleth(filtered_dengue_case_counts, 'August')
```
